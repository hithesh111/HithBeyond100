Sampling from Distributions using Universal Law of Standard Uniform</h2>
Wrote code to generate and visualize a random sample from a distribution with a known CDF by sampling from a Standard Uniform Distribution and applying Universal Law.<br />
<br />
The law states that if F is the cdf of the distribution from which we want to draw samples. Then X = F-1(U) has cdf F.<br>
    (Read F-1 as 'F inverse'. I'll figure out how to latex soon.)<br />
<br />
We generate a bunch of numbers from Uniform(0,1) distribution and then making the F-1 transformation on these numbers to get random numbers from the distribution we desire. I've illustrated it using plots.<br />

```python
#Importing required packages
import numpy as np
import math
import seaborn as sns
```

```python
#Creates a list of 10000 random numbers from Uniform(0,1)
X = np.random.rand(10000)
sns.distplot(X)
```

```python
#Calculates the inverse function of the cdf of Exponential and Geometric
#(Inverses of other distributions can be very complicated.)
def exp_inv_cdf(X,par):
    return [-math.log(1-x)/(par) for x in X]

def geom_inv_cdf(X,p):
    return [math.floor(math.log(1-x)/math.log(1-p)) for x in X]
```

```python
#The createed exponential sample with parameter 0.1
sns.distplot(exp_inv_cdf(X,0.1),kde=False)
```

```python
#Created Geometric Sample with parameter 1/2
#As expected probability of X  taking value 0 is around 5000/10000, probability of X taking value 1 is around2500/10000 etc
sns.distplot(geom_inv_cdf(X,0.5),kde = False)
```
